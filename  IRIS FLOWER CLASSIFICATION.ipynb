## Seggregate `Input` and `outputs`

```python
df.columns
```

```python
X = df.drop(columns="species")
```

```python
y = df['species']
```

## Split the data into `TRAIN` and `TEST`

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print(X_train.shape,y_train.shape)
```

```python
print(X_test.shape,y_test.shape)
```

## Data preparation

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_transformed = pd.DataFrame(scaler.fit_transform(X_train),
                                  columns=scaler.get_feature_names_out(),index = X_train.index)
```

```python
X_test_transformed = pd.DataFrame(scaler.transform(X_test),columns=scaler.get_feature_names_out(),
                                 index = X_test.index)
```

## Building the `MODEL`

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
%%time
reg.fit(X_train_transformed,y_train)
```

```python
y_train_pred = reg.predict(X_train_transformed)
```

```python
y_test_pred = reg.predict(X_test_transformed)
```

```python
from sklearn import metrics
```

```python
log_train_score = metrics.accuracy_score(y_train_pred,y_train)
print(log_train_score)
```

```python
log_test_score = metrics.accuracy_score(y_test_pred,y_test)
print(log_test_score)
```

```python
print(metrics.classification_report(y_train_pred,y_train))
```

```python
print(metrics.classification_report(y_test_pred,y_test))
```
